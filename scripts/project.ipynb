```python
# import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset and display the first five rows of the dataset data.csv
df = pd.read_csv('data.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# convert the datatype of the columns
df.State = df.State.convert_dtypes()
df.Indicator = df.Indicator.convert_dtypes()
df.Group= df.Group.convert_dtypes()
df.Subgroup = df.Subgroup.convert_dtypes()
df.Time_Period_Start_Date = pd.to_datetime(df.Time_Period_Start_Date)
df.Time_Period_End_Date = pd.to_datetime(df.Time_Period_End_Date)
```

```python
# plot a graph to show what indicators are there by education 
sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
sns.countplot(x='Indicator', hue='Group', data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
# plot a graph to show what indicators are there by education
plt.figure(figsize=(10,6))
sns.countplot(x='Indicator', hue='Group', data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
# access the data of group 'By Education'
df1 = df[df['Group'] == 'By_Education']
df1.head()
```

```python
#add palette to the graph to make it more colorful and attractive and background color to white
sns.set(style="white")
plt.figure(figsize=(10,6))
sns.countplot(x='Indicator', hue='Subgroup', data=df1, palette='Set1')
plt.xticks(rotation=90)
plt.title('Indicators by Subgroup')
plt.xlabel('Indicator')
plt.ylabel('Subgroup')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()


```

```python
# plot a graph to show what indicators are there by education and mean of the values
plt.figure(figsize=(10,6))
sns.barplot(x='Indicator', y='Value', hue='Subgroup', data=df1, palette='Set1')
plt.xticks(rotation=90)
plt.title('Indicators by Subgroup')
plt.xlabel('Indicator')
plt.ylabel('Value')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
from vega_datasets import data

states = alt.topo_feature(data.us_10m.url, 'states')

alt.Chart(states).mark_geoshape().encode(
    color='Value:Q'
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(df2, 'State', ['Value'])
).project(
    type='albersUsa'
).properties(
    width=500,
    height=300
)
```
